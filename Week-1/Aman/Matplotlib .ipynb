```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [1,4,9]
z = [10,5,0]
plt.plot(x,y)
plt.plot(x,z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.show()
```

## Add legend

```python
plt.plot(x,y,label ="y")
plt.plot(x,z,label = "z")
plt.legend(loc='best')#for location
plt.show()
```

## Increase fig. size and resolution

```python
plt.figure(figsize = (8,3), dpi=180) #dots per inches(dpi)
plt.plot(x,y,label ="y")
plt.plot(x,z,label = "z")
plt.legend(loc='best')

plt.show()
```

## Adding grid

```python
plt.figure(figsize = (8,3), dpi=180) #dots per inches(dpi)
plt.plot(x,y,label ="y")
plt.plot(x,z,label = "z")
plt.legend(loc='best')
plt.grid(True)

plt.show()
```

## spines

```python
plt.figure(figsize = (8,3), dpi=180) #dots per inches(dpi)
plt.plot(x,y,label ="y")
plt.plot(x,z,label = "z")
plt.legend(loc='best')
plt.grid(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)


plt.show()
```

## Histogram

```python
y = np.random.randn(1000)
plt.hist(y)
plt.show
```

```python
plt.hist(y,bins=20,color='r',edgecolor='b',normed=True)
plt.show
```

```python
plt.hist(y,bins=20,color='r',edgecolor='b',cumulative=True)
plt.show
```

## Bar charts

```python
plt.bar([1, 2, 3], [30, 40, 50]);
plt.show()
```

## pie charts

```python
x = [50,45,30]
```

```python
plt.figure(figsize=(5,5))
a = ['cars','bikes','buses']
plt.pie(x,labels=a)
plt.show()
```

```python
plt.figure(figsize=(8,8))
x = [4, 9, 21, 55, 30, 18]
labels = ['bmw', 'audi', 'ferrari', 'ford', 'honda','jaguar']
explode = [0.6, 0.1, 0, 0, 0.1, 0]
plt.pie(x, labels=labels, explode=explode,autopct='%1.1f%%')#autopct=for write the values
plt.show()
```

## Scatter plots

```python
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y);
plt.show()
```

```python
size = 50*np.random.randn(100)
colors = np.random.rand(100)
plt.scatter(x, y, s=size, c=colors);
plt.show();
```

## Subplots

```python
fig = plt.figure(figsize=(10,10))

ax1 = fig.add_subplot(221)
ax1.hist(y);

ax2 = fig.add_subplot(222)
ax2.bar([1, 2, 3], [30, 40, 50]);

ax3 = fig.add_subplot(223)
x = [45, 35, 20]
labels = ['Cars', 'bikes', 'buses']
ax3.pie(x, labels=labels);

ax4 = fig.add_subplot(224)
x = [4, 9, 21, 55, 30, 18]
labels = ['bmw', 'audi', 'ferrari', 'ford', 'honda','jaguar']
explode = [0.2, 0.1, 0, 0, 0.2, 0]
ax4.pie(x, labels=labels, explode=explode, autopct='%1.1f%%');

plt.savefig('aman hooda', dpi=200)


plt.show()
```

## Contour plots

```python
xlist =np.linspace(-1,1,3)#start from -1 to 1 in 3 steps
ylist = np.linspace(-1,1,3)
x,y=np.meshgrid(xlist,ylist)
z=np.sqrt(x**2+y**2)
cp = plt.contour(x,y,z)
plt.clabel(cp,inline=True,fontsize=10)
plt.contourf(x,y,z)#for coloring
```
